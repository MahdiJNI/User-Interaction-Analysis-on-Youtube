Converting values with unit to only numeric values from view_1 to view_20

```python
col_list_view_ = []
for col_view_ in df:
    col_list_view_.append(col_view_)
col_list_view_ = col_list_view_[6:]


for num_view_ in range(0, len(col_list_view_), 4):
    for data_view_ in df[col_list_view_[num_view_]]:
        edition_view_ = df[col_list_view_[num_view_]].copy()
        #print(col_list_view_[num_view_])
        for index_view_, row_view_ in edition_view_.items():
            if 'm' in row_view_:
                edition_view_.loc[index_view_] = int(float(edition_view_.loc[index_view_].split('m')[0]) * 1000000)
                
            if 'k' in row_view_:
                edition_view_.loc[index_view_] = int(float(edition_view_.loc[index_view_].split('k')[0]) * 1000)
            # if 'M' not in row_view_ and 'K' not in row_view_ and 'm' not in row_view_ and 'k' not in row_view_ :
            #     print(edition_view_.loc[index_view_], col_list_view_[num_view_])
    df[col_list_view_[num_view_]] = edition_view_
```
