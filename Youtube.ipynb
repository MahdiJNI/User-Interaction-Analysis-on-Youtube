Changing v,d,l,c to video, duration, like, comment :

```python
for column in df:
    if len(column) < 4 :
        if len(column.split('v')) == 2:
            new = (str(column.split('v')[0]) + "view_") + column.split('v')[1]
            df.rename(columns={column:new.strip()}, inplace=True)
        elif len(column.split('d')) == 2:
            new = (str(column.split('d')[0]) + "duration_") + column.split('d')[1]
            df.rename(columns={column:new.strip()}, inplace=True)
        elif len(column.split('l')) == 2:
            new = (str(column.split('l')[0]) + "like_") + column.split('l')[1]
            df.rename(columns={column:new.strip()}, inplace=True)
        elif len(column.split('c')) == 2:
            new = (str(column.split('c')[0]) + "comment_") + column.split('c')[1]
            df.rename(columns={column:new.strip()}, inplace=True)
            
```
